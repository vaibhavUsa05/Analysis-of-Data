```python
# Q. 2) In which year the highest number of the TV Shows & Movies were released ? Show with Bar Graph.
```

```python
df["Release_Date"] = df["Release_Date"].str.strip() 
df["year"]=pd.to_datetime(df["Release_Date"])
```

```python
df.head()
```

```python
date_year=df["year"].dt.year.value_counts()
date_year
```

```python
plt.figure(figsize=(8,6))
yaxis=date_year.values.tolist()
xaxis=date_year.index.tolist()
plt.bar(xaxis,yaxis)
plt.title("Year wise movie counts")
plt.xlabel("Year")
plt.ylabel("Total cpounts of movies and shows")
plt.show()
```

```python
#  How many Movies & TV Shows are in the dataset ? Show with Bar Graph.
```

```python
count_of_movies=(df["Category"]=="Movie").value_counts()
count_of_movies=count_of_movies.values[0]
count_of_tvshows=(df["Category"]=="TV Show").value_counts().values[1]
```

```python
categories = df.groupby("Category")['Category'].count()
categories_shows=categories.index.tolist()
categories_shows_labels=categories.values.tolist()
plt.bar(categories_shows,categories_shows_labels,color="green",width=0.7)
for i, value in enumerate(categories_shows_labels):
    plt.text(i, value, str(value), ha='center', va='bottom')

plt.title("Count of ALL MOVIES AND TV shows")
plt.xlabel("Categories")
plt.ylabel("Count of all categories")
plt.show()
```

```python
# Show Top 10 Directors, who gave the highest number of TV Shows & Movies to Netflix ?
```

```python
plt.figure(figsize=(14,6))
df_director=df.groupby("Director")["Director"].count()
df_director=df_director.sort_values(ascending=False)
df_director=df_director.head(10)
director_name=df_director.index.tolist()
director_creations=df_director.values.tolist()
plt.bar(director_name,director_creations,color="orange")
plt.title("The Directors and their creations")
plt.xlabel("Director name")
plt.ylabel("Count of all categories")
plt.show()
```
